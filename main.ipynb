## Step 4: Modelling

### divide data to training data and testing data

```python
# df.to_csv('data/Cleaned_Data_exported.csv', index=False)

df = pd.read_csv(path + 'Cleaned_Data.csv')

print(df.shape)

test_data = df.sample(n=200, random_state=0)
train_data = df.drop(test_data.index).sample(n=800, random_state=0)

print(train_data.shape)
print(test_data.shape)

print(train_data[:10].to_string())
print(test_data[:10].to_string())
```

### 1. KNN



*   Pros: Simple to implement and understand; works well with a small number of input variables.

*   Cons: Computationally expensive, especially with large datasets; performance degrades with high dimensionality.



```python
import numpy as np
from collections import Counter



# implement knn

def euclidean_distance(row1, row2):
    """
    A function to calculate the Euclidean distance between two points.
    """
    # print(type(row1), type(row2))
    # print(row2[:5])
    distance = 0.0
    decimal_keys = ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'duration_ms']
    for i in decimal_keys:
        row1_val = float(row1.loc[i])
        row2_val = float(row2.loc[i])
        distance += (row1_val - row2_val)**2
    return np.sqrt(distance)

def calculate_accuracy(predictions):
    correct_predictions = 0

    for actual, predicted in predictions:
        if actual == predicted:
            correct_predictions += 1

    total_predictions = len(predictions)
    accuracy = (correct_predictions / total_predictions)
    return accuracy


def knn_predict(train_data, test_row, num_neighbors):
    """
    A function to predict the class of a test row using kNN algorithm.
    """
    distances = []

    # Calculate distances from test_row to all training data
    for i in range(len(train_data)):
        dist = euclidean_distance(train_data.iloc[i, :], test_row)
        distances.append((train_data.index[i], dist))

    distances.sort(key=lambda tup: tup[1])
    nearest_neighbors = distances[:num_neighbors]

    # Extract the class labels of the nearest neighbors
    output_values = [df.iloc[row[0], 18] for row in nearest_neighbors]

    # Return the most common class label
    # print(Counter(output_values).most_common(1))
    # (actual genre, predict genre)
    prediction = (Counter(output_values).most_common(1)[0][0], test_row.playlist_genre)
    return prediction

# Example usage:
# Assume we have a dataset 'dataset' where each row is a data point and
# the last column is the class label.
# Also assume 'test_data' is a list of rows without class labels.

# Predict the class for each test instance
k = 60  # Number of neighbors
knn_predictions = [knn_predict(train_data, test_instance, k) for _, test_instance in test_data.iterrows()]

knn_score = calculate_accuracy(knn_predictions)

print("predict accuracy: ", knn_score)
```

### 2. SVM

* Pros: Effective in high-dimensional spaces; works well with a clear margin of separation and is robust against overfitting.

* Cons: Requires careful tuning of parameters; not suitable for larger datasets as training time can be long.

```python
## SVM

from sklearn import svm
from sklearn.metrics import accuracy_score

# Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

# Train the model using the training sets
clf.fit(train_data.iloc[:, 3:13], train_data.iloc[:, 18])

# Predict the response for test dataset
y_pred = clf.predict(test_data.iloc[:, 3:13])

svm_score = accuracy_score(test_data.iloc[:, 18], y_pred)
print("Accuracy:", svm_score)

```

### 3. Decision Tree

* Pros: Easy to interpret and visualize; can handle both numerical and categorical data.

* Cons: Prone to overfitting, especially with complex trees; can be unstable with small variations in data.

```python
## Decision Tree

from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the model using the training sets
clf.fit(train_data.iloc[:, 3:13], train_data.iloc[:, 18])

# Predict the response for test dataset
y_pred = clf.predict(test_data.iloc[:, 3:13])

deci_tree_score = accuracy_score(test_data.iloc[:, 18], y_pred)
print("Accuracy:", deci_tree_score)
```

### 4. Random Forest

* Pros: Reduces overfitting compared to decision trees; highly accurate and versatile for both classification and regression tasks.

* Cons: Less interpretable than a single decision tree; can be computationally intensive and slow to train.

```python
## Random Forest

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)
clf.fit(train_data.iloc[:, 3:13], train_data.iloc[:, 18])

y_pred = clf.predict(test_data.iloc[:, 3:13])

random_forest_score = accuracy_score(test_data.iloc[:, 18], y_pred)


print("Accuracy: ", random_forest_score)
```

### 5. Gradient Boosting

* Pros: Often provides predictive accuracy that cannot be beaten; flexible with various loss functions.

* Cons: Can overfit on noisy datasets; requires careful tuning of parameters and is computationally expensive.

```python
## Gradient Boosting

from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier()
clf.fit(train_data.iloc[:, 3:13], train_data.iloc[:, 18])

y_pred = clf.predict(test_data.iloc[:, 3:13])
grad_boost_score = accuracy_score(test_data.iloc[:, 18], y_pred)
print("Accuracy: ", grad_boost_score)
```

## Step 5: Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

# Parameter grid
param_grid_rf = {'n_estimators': [100, 200, 300, 400, 500, 600], 'max_depth': [10, 15, 20, 25, 30]}

# Grid search
grid_search_rf_random_forest = GridSearchCV(RandomForestClassifier(), param_grid_rf, cv=5)
grid_search_rf_random_forest.fit(train_data.iloc[:, 3:13], train_data.iloc[:, 18])

# Best parameters and performance
print("Best parameters for Random Forest:", grid_search_rf_random_forest.best_params_)
print("Best score:", grid_search_rf_random_forest.best_score_)

```
